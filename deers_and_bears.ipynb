```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df_bear = pd.read_csv("data/bear.csv")
df_deer = pd.read_csv("data/deer.csv")
```

```python
fig, axs = plt.subplots(nrows=len(df_bear.columns), figsize=(20, 40))
fig.tight_layout(pad=5.0)

columns = df_bear.columns
for index, column in enumerate(columns):
    concatenated = pd.concat([df_bear.assign(dataset='Bears'), df_deer.assign(dataset='Deers')]).reset_index().rename(columns={ "index": "Time" })  
    sns.lineplot(x="Time", y=column, data=concatenated, hue="dataset", ax=axs[index])
```
