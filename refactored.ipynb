```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_theme()
```

```python
df = pd.read_csv("data/deer-plants-refactored.csv")
```

```python
columns = ["population", "speed", "size", "births", "deathsByHunger", "deathsByThirst", "deathsByAge", "deathsTotal"]
fig, axs = plt.subplots(nrows=len(columns), figsize=(20, 7 * len(columns)))
fig.tight_layout(pad=5.0)

for index, column in enumerate(columns): 
    g = sns.lineplot(x="time", y=column, data=df, hue="species", ax=axs[index])
    g.axes.set_ylim(0)
    g.axes.set_xlim(0)
```

```python
df_deers = df[df['species'] == "Deer"]
df_grass = df[df['species'] == "Grass"]

np.corrcoef(df_deers['population'], df_grass['population'])
```

```python
plt.figure(figsize=(20, 7))
plt.title("Populations against Time")
g = sns.lineplot(data=df[(df['species'] == "Deer") | (df['species'] == "Grass")], x="time", y="population", hue="species", palette=["b", "g"], linewidth=2.5)
g.axes.set_ylim(0)
g.axes.set_xlim(0)

fig = plt.figure(figsize=(20, 7))
plt.title("Speed against Time")
g = sns.lineplot(x="time", y="speed", data=df[(df['species'] == "Deer")], color="b", linewidth=2.5)
g.axes.set_ylim(0)
g.axes.set_xlim(0)

fig = plt.figure(figsize=(20, 7))
plt.title("Size against Time")
g = sns.lineplot(x="time", y="size", data=df[(df['species'] == "Deer")], color="orange", linewidth=2.5)
g.axes.set_ylim(0)
g.axes.set_xlim(0)

fig = plt.figure(figsize=(20, 7))
plt.title("Speed/Size against Time")
g = sns.lineplot(x="time", y="value", hue="variable", data=pd.melt(df[(df['species'] == "Deer")][["time", "size", "speed"]], ["time"]), linewidth=2.5)
g.axes.set_ylim(0)
g.axes.set_xlim(0)
```
