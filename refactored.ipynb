```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_theme()
```

```python
df = pd.read_csv("data/deer_with_plants.csv")
```

```python
columns = [col for col in df.columns if col not in ["species", "time"]]
fig, axs = plt.subplots(nrows=len(columns), figsize=(20, 7 * len(columns)))
fig.tight_layout(pad=5.0)

for index, column in enumerate(columns): 
    g = sns.lineplot(x="time", y=column, data=df, hue="species", ax=axs[index])
    g.axes.set_ylim(0)
    g.axes.set_xlim(0)
```

```python
plt.figure(figsize=(20, 10))
sns.lineplot(data=df[(df['species'] == "Deer") | (df['species'] == "Grass")], x="time", y="population", hue="species")
```

```python
df_deers = df[df['species'] == "Deer"]
df_grass = df[df['species'] == "Grass"]
```
